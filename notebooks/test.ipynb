```python
def make_square(im, pad_val):
    (a, b) = im.shape[0:2]
    if a > b:
        im_size = ((0,0),(0,a-b))
    else:
        im_size = ((0,b-a),(0,0))
    return np.pad(im, im_size, mode = 'constant', constant_values = pad_val)
```
